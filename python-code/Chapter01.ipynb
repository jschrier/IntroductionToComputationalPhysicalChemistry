# 1.1 BASIC DEFINITIONS

## 1.1.3 Python Implementation

```python
import numpy as np
import matplotlib.pyplot as plt



def psi1DPIB(n, x, L):
    return np.sqrt(2/L)*np.sin(n * np.pi * x / L)

def energy1DPIB(n, L, m:1, hbar:1):
    return (hbar**2 * np.pi**2 * n**2) / (2 * m * L**2)

# assume default mass of 1 me and atomic units of hbar = 1
L  = 5 # set L = 5 a0, i.e., units of Bohr radii, a0
x_range = np.linspace(0, L, 100) # need alternative to range
psi1_values = [psi1DPIB(1, x, L) for x in x_range]
psi2_values = [psi1DPIB(2, x, L) for x in x_range]

# Plot the function
plt.plot(x_range, psi1_values )
plt.plot(x_range, psi2_values)

# Add axis labels
plt.xlabel('x')
plt.ylabel('psi(x)')
#plt.axes().set_frame_on(True)
#plt.gca().set_frame_on(True)

# Show the plot
plt.show()
```

```python

prob1_values = [psi1DPIB(1, x, L)*psi1DPIB(1, x, L) for x in x_range]
prob2_values = [psi1DPIB(2, x, L)*psi1DPIB(2, x, L) for x in x_range]

plt.plot(x_range, prob1_values, x_range, prob2_values)
plt.show()
```

```python
import sympy as sp

def symbolic_psi1PIB(n):
    x = sp.Symbol('x')
    L = sp.Symbol('L')
    return sp.sqrt(2/L)*sp.sin(n * sp.pi * x / L)

psi1 = symbolic_psi1PIB(1)
psi2 = symbolic_psi1PIB(2)

x = sp.Symbol('x')
integral = sp.integrate(psi1*psi1, x)
integral
```

# 1.4 DEGENERACY: THE 2D PARTICLE IN A BOX

```python
def psi2DPIB(nx, ny, x, y, Lx, Ly):
    return psi1DPIB(nx, x, Lx)*psi1DPIB(ny, y, Ly)

def energy2DPIB(nx, ny, Lx, Ly, m:1, hbar:1):
    return energy1DPIB(nx, Lx, m, hbar) + energy1DPIB(ny, Ly, m, hbar)

L = 5
x_range = np.linspace(0, L, 100)
psi2d_values = [psi2DPIB(1, 2, x, y, L, L) for x in x_range for y in x_range]
```

# 1.5 VISUALIZING PROBABILITY DENSITIES

```python
def psiSquared1DPIB(n, x, L):
    return psi1DPIB(n,x,L) * psi1DPIB(n, x, L)
```

# 1.5.2 Implementation in Mathematica

```python
accepted_points = []
L = 5
for i in range(10000):
    x = np.random.uniform(0,5)
    if ( psiSquared1DPIB(1, x, L) >= np.random.uniform(0, 0.5) ):
        accepted_points.append(x)

plt.hist(accepted_points)
plt.show()
```
